### 阈值处理
#### 图像阈值

#### ret, dst = cv2.threshold(src, thresh, maxval, type)

- src： 输入图，只能输入单通道图像，通常来说为灰度图
- dst： 输出图
- thresh： 阈值
- maxval： 当像素值超过了阈值（或者小于阈值，根据type来决定），所赋予的值
- type：二值化操作的类型，包含以下5种类型： cv2.THRESH_BINARY； cv2.THRESH_BINARY_INV； cv2.THRESH_TRUNC； cv2.THRESH_TOZERO；cv2.THRESH_TOZERO_INV

- cv2.THRESH_BINARY           超过阈值部分取maxval（最大值），否则取0
- cv2.THRESH_BINARY_INV    THRESH_BINARY的反转
- cv2.THRESH_TRUNC            大于阈值部分设为阈值，否则不变
- cv2.THRESH_TOZERO          大于阈值部分不改变，否则设为0
- cv2.THRESH_TOZERO_INV  THRESH_TOZERO的反转

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img=cv.imread('data/cat.jpg')
img_gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
ret,thresh1 = cv.threshold(img_gray,127,255,cv.THRESH_BINARY)
# 超过阈值取最大值，否则为0
ret,thresh2 = cv.threshold(img_gray,127,255,cv.THRESH_BINARY_INV)
# 反转
ret,thresh3 = cv.threshold(img_gray,127,255,cv.THRESH_TRUNC)
# 大于阈值为阈值，否则不变
ret,thresh4 = cv.threshold(img_gray,127,255,cv.THRESH_TOZERO)
# 大于阈值部分不变，否则为0
ret,thresh5 = cv.threshold(img_gray,127,255,cv.THRESH_TOZERO_INV)
# 反转
titles = ['Original','BINARY','BINARY_INV','TRUNC','OZERO','TOZERO_INV']
images = [img,thresh1,thresh2,thresh3,thresh4,thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

### 平滑处理
滤波操作：均值，方框，高斯，中值

```python
# 滤波前
img= cv.imread('data\lenaNoise.png')
cv.imshow('img',img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 均值滤波 简单平均卷积操作
blur = cv.blur(img,(3,3))
cv.imshow('blur',blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 方波滤波 简单平均卷积操作
box = cv.boxFilter(img,-1,(3,3),normalize=True)
cv.imshow('box',box)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 方波滤波 简单平均卷积操作
box = cv.boxFilter(img,-1,(3,3),normalize=False)
cv.imshow('box',box)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 高斯滤波
aussian = cv.GaussianBlur(img, (5, 5), 1)  

cv.imshow('aussian', aussian)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 中值滤波
# 相当于用中值代替
median = cv.medianBlur(img, 5)  # 中值滤波

cv.imshow('median', median)
cv.waitKey(0)
cv.destroyAllWindows()
# 最好的去除点的滤波操作
```

```python
# 展示所有的
res = np.hstack((blur,aussian,median))
#print (res)
cv.imshow('median vs average', res)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# 展示所有的
res = np.vstack((blur,aussian,median))
#print (res)
cv.imshow('median vs average', res)
cv.waitKey(0)
cv.destroyAllWindows()
```
